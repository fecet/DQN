```python
import numpy as np
import pandas as pd
data = np.load('data/trajectories.npy', allow_pickle=True)
data[0]
import utils
one_step = utils.get_steps(data,1,1)[-1]
```

```python
input_array = one_step[0][0].T
input_array
```

```python
input_array.T
```

```python
def find_in_array(array):
    for idx,num in enumerate(array):
        if num == 1:
            return idx

cards = []
for card in input_array:
    cards.append(find_in_array(card))

cards
```
